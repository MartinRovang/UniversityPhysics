```python
import numpy as np; np.random.seed(8)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
plt.figure(figsize=(15,10))
x = 1
j = 0
k = 0
z = 500
R = []
i = 0
while k< 300:
    y =(x*0.10*z)/(j+z)
    R.append(y)
    j = x + j
    k = y + k
    x = np.random.randint(50,1000)
    i = 1 + i
print(i)
z1 = np.linspace(1,len(R),len(R))
z2 = np.array(R)
ax = sns.regplot(x=z1, y=z2, color="r",logx=True, truncate=True)
plt.show()
```
