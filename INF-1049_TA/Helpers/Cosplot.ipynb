```python
#Importerer ndvendige biblioteker 
import matplotlib.pyplot as plt
import numpy as np

#Lager et intervall for tid. 0-10 sekunder med 300 "steg/delta t" for  jevne ut kurven
t = np.linspace(0, 10, 300)

#lager to cosinuskurver med periode T=2 og T=3
T1 = 2
x1 = np.cos(2*np.pi*t/T1)

T2 = 3
x2 = np.cos(2*np.pi*(t/T2))

#Plotter kurvene med definisjonsmengden t og verdimengden til cosinuskurvene
plt.plot(t,x1, label='x1')
plt.plot(t,x2, label='x2')
plt.legend(loc='upper left')

#Vis kurve.
plt.show()
```

```python
#Importerer ndvendige biblioteker 
import matplotlib.pyplot as plt
import numpy as np

#Lager et intervall for tid. 0-10 sekunder med 300 "steg/delta t" for  jevne ut kurven
t = np.linspace(0, 10, 80)

#Cosinuskurver med periode T=2
T1 = 2
x1 = np.cos((2*np.pi*t)/T1)

#Lager array for med verdiene til alle differansene til x1 og t og deler det med hverandre
el2 = []
#For lkke for  fylle opp array.
for i, el in enumerate(x1):
    first = x1[i]
    second = x1[i - 1]
    fourth = t[i - 1]
    third = t[i]
    x2 = (first - second)
    t2 = (third - fourth)
    ans = x2/t2
    #Legg inn i array.
    el2.append(ans)

#Eksakt derivert
Ex1 = (-2*np.pi*np.sin((2*np.pi)*t/T1)/T1)


plt.plot(t,el2, label='x1')
plt.plot(t,Ex1, label='Ex1')
plt.legend(loc='upper left')


#Vis kurve.
plt.show()
```

```python

# countour map -- topographical data, or anything in 3 dimensions

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
plt.show()
```
