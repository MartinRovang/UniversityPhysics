```python
#http://tutorial.math.lamar.edu/Classes/DE/EulersMethod.aspx


import numpy as np
import matplotlib.pyplot as plt

i = 0
dt = 0.01
k = 1
t = 0
m = 1
dx = 0
x = 1
ddxarray = []
dxarray = []
xarray = [1]
tarray = []

while t < 10:
    ddx = -k*xarray[i]/m
    ddxarray.append(ddx)
    dx += ddxarray[i]*dt
    dxarray.append(dx)
    x += dxarray[i]*dt
    xarray.append(x)
    t += dt
    tarray.append(t)
    i += 1

plt.figure(figsize = (15,10))
plt.subplot(111)
plt.plot(tarray, ddxarray, label = "Akselerasjon")
plt.subplot(111)
plt.plot(tarray, dxarray, label = "Hastighet")
plt.subplot(111)
plt.plot(tarray, xarray[:-1], label = "Posisjon")
plt.legend(loc = "best")
plt.show()
```
