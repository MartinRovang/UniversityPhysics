```python
from math import sin, pi
import numpy as np


def f1(x):
    return np.sin(x) + 1

def f2(x):
    return x**2


def integrate(f, a ,b ,n = 100):
    k = np.array([i for i in range(n-1)])
    return ((b-a)/n)*((f(a)/2)+np.sum(f(a+k*(b-a)/n))+f(b)/2)



print("The integral of f1 in the interval [0,pi] = %.2f"%(integrate(f1, 0, pi)))
print("The integral of f1 in the interval [-2,2] = %.2f"%(integrate(f2, -2, 2)))


```

```python
class polynomial(object):
    def __init__(self, coefficients):
        self.coeff = coefficients #Hente ut det som er inni listen
    
        
    def __call__(self, x):
        s = 0
        for i in range(len(self.coeff)):
            s += self.coeff[i]*x**i
        return s
        
    def __add__(self, other):
        if len(self.coeff) > len(other.coeff):
            result_coeff = self.coeff[:]
            for i in range(len(other.coeff)):
                result_coeff[i] += other.coeff[i]
        else:
            result_coeff = other.coeff[:]
            for i in range(len(self.coeff)):
                result_coeff[i] += self.coeff[i]
        return polynomial(result_coeff)
    

    def __str__(self):
        s = ''
        for i in range(0, len(self.coeff)):
            if self.coeff[i] != 0:
                s += ' + %g*x^%d' %(self.coeff[i], i)
        #fix layout
        s = s.replace('+ -','- ')
        s = s.replace('x^0', '1')
        s = s.replace(' 1*', ' ')
        s = s.replace('*1', '')
        s = s.replace('x^1 ', 'x ')
        if s[0:3] == ' + ':
            s = s[3:]
        if s[0:3] == ' - ':
            s = '-' + s[3:]
        return s

    
p1 = polynomial([7, 0, 3, 0, -5, 0, 1])
print(p1)
p2 = polynomial([1, 1, 2, 3, 0, 2, -5, 1])
print(p2)
p3 = p1 + p2
print(p3)
```
